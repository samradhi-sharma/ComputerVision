```python
# we are working only with rgb color spaces in rgb coding means combination of red, green, blue
# there are two more models HSL(hue, saturation, lightness)
# HSV(hue, saturation, value)
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog1.jpg')
```

```python
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img) 
```
