```python
import cv2
import numpy as np

# Callback function for mouse events
def create_circle(event, x, y, flags, param):
    global img  # Ensure we're modifying the global image
    if event == cv2.EVENT_RBUTTONDOWN:  # Right-click to draw a circle
        cv2.circle(img, (x, y), 100, (0, 0, 255), thickness=10)
        cv2.imshow('dog', img)  # Refresh the window with the updated image

# Load the image
img = cv2.imread("DATA/dog2.jpg")  # Ensure the path is correct

cv2.namedWindow('dog')
cv2.setMouseCallback('dog', create_circle)

while True:
    cv2.imshow('dog', img)  # Show the image
    if cv2.waitKey(20) & 0xFF == 27:  # Press 'Esc' to exit
        break

cv2.destroyAllWindows()
```
