```python
# IMAGE THRESHOLDING
# segmenting a image into different parts, will convert an image to consist of only two values
# white or black
```

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/rainbow.jpg',0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
img.max()
```

```python
ret, thresh1=cv2.threshold(img, 127,255,cv2.THRESH_TOZERO_INV)
# it simply means any value below 127 will be shifted down to zero and 
# any value above 255 will be shifted max up to 255
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
img = cv2.imread('DATA/crossword.png',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax=fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img,200,255,cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)
show_pic(th2)
```

```python
blended =cv2.addWeighted(src1=th1,alpha=0.6,src2=th2,beta=0.4,gamma=0)
show_pic(blended)
```
